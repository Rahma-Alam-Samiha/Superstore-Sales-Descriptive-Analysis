```python
import pandas as pd

def search_value(dataframe, column_name, search_value):
    """
    Search for a value in a specific column of a DataFrame.

    Parameters:
    - dataframe: Pandas DataFrame
    - column_name: Name of the column to search
    - search_value: Value to search for

    Returns:
    - True if the value is found, False otherwise
    """
    if column_name not in dataframe.columns:
        print(f"Column '{column_name}' not found in the DataFrame.")
        return False

    found = search_value in dataframe[column_name].values

    if found:
        print(f"Value '{search_value}' found in column '{column_name}'.")
        return True
    else:
        print(f"Value '{search_value}' not found in column '{column_name}'.")
        return False

# Example usage:
# Assuming you have a DataFrame named 'df' with a column named 'YourColumn'
# and you want to search for the value 'YourValue'
# Replace 'YourColumn', 'YourValue', and 'df' with your actual column name, value, and DataFrame.

search_value(data, 'Product Name', False)
```

```python
def count_occurrences(dataframe, column_name, search_value):
    """
    Count how many times a value is present in a specific column of a DataFrame.

    Parameters:
    - dataframe: Pandas DataFrame
    - column_name: Name of the column to search
    - search_value: Value to count occurrences for

    Returns:
    - Number of occurrences of the value in the specified column
    """
    if column_name not in dataframe.columns:
        print(f"Column '{column_name}' not found in the DataFrame.")
        return 0

    count = dataframe[column_name].eq(search_value).sum()

    print(f"Value '{search_value}' occurs {count} times in column '{column_name}'.")
    return count

# Example usage:
# Assuming you have a DataFrame named 'df' with a column named 'YourColumn'
# and you want to count occurrences of the value 'YourValue'
# Replace 'YourColumn', 'YourValue', and 'df' with your actual column name, value, and DataFrame.

count = count_occurrences(data, 'Product Name', False)
```

```python
def remove_value(dataframe, column_name, remove_value):
    """
    Remove rows containing a specific value in a specific column of a DataFrame.

    Parameters:
    - dataframe: Pandas DataFrame
    - column_name: Name of the column to search
    - remove_value: Value to be removed

    Returns:
    - DataFrame with rows containing the specified value removed
    """
    if column_name not in dataframe.columns:
        print(f"Column '{column_name}' not found in the DataFrame.")
        return dataframe

    new_dataframe = dataframe[dataframe[column_name] != remove_value]

    print(f"Rows with value '{remove_value}' removed from column '{column_name}'.")
    return new_dataframe

# Example usage:
# Assuming you have a DataFrame named 'df' with a column named 'YourColumn'
# and you want to remove rows containing the value 'YourValue'
# Replace 'YourColumn', 'YourValue', and 'df' with your actual column name, value, and DataFrame.

new_df = remove_value(data, 'Product Name', False)
```

```python
def detect_outliers(dataframe, column_name, threshold=2):
    """
    Detect outliers in a specific column of a DataFrame using the Z-score method.

    Parameters:
    - dataframe: Pandas DataFrame
    - column_name: Name of the column to check for outliers
    - threshold: Z-score threshold for outlier detection (default is 3)

    Returns:
    - DataFrame with an additional column indicating whether each row is an outlier
    """
    if column_name not in dataframe.columns:
        print(f"Column '{column_name}' not found in the DataFrame.")
        return dataframe

    # Calculate Z-scores
    z_scores = np.abs((dataframe[column_name] - dataframe[column_name].mean()) / dataframe[column_name].std())

    # Identify outliers based on the threshold
    dataframe['IsOutlier'] = z_scores > threshold

    print(f"Outliers detected in column '{column_name}' using Z-score method.")
    return dataframe

# Example usage:
# Assuming you have a DataFrame named 'df' with a column named 'YourColumn'
# Replace 'YourColumn' and 'df' with your actual column name and DataFrame.
```

```python
from sklearn.ensemble import IsolationForest

def remove_outliers_isolation_forest(df, columns, contamination=0.5, random_state=42):
    """
    Remove outliers from specific columns of a DataFrame using the Isolation Forest algorithm.

    Parameters:
    - df: DataFrame
    - columns: List of column names to consider for outlier removal
    - contamination: The proportion of outliers in the data (default is 0.05)
    - random_state: Seed for reproducibility (default is 42)

    Returns:
    - DataFrame without outliers
    """
    outlier_indices = []

    for column in columns:
        # Create an Isolation Forest model
        model = IsolationForest(contamination=contamination, random_state=random_state)

        # Fit the model to the column data
        model.fit(df[column].values.reshape(-1, 1))

        # Predict outliers (1 for inliers, -1 for outliers)
        predictions = model.predict(df[column].values.reshape(-1, 1))

        # Identify indices of outliers for the current column
        outliers = df[predictions == -1].index
        outlier_indices.extend(outliers)

    # Remove rows containing outliers
    df_no_outliers = df.drop(outlier_indices)

    return df_no_outliers

# Usage example:
# Specify the columns for outlier removal
columns_to_check = ['Sales', 'Quantity', 'Discount', 'Profit']

# Remove outliers using the Isolation Forest algorithm

```

```python
new_df = remove_outliers_isolation_forest(new_df, columns=columns_to_check, contamination=0.5, random_state=50)
new_df = remove_outliers_isolation_forest(new_df, columns=columns_to_check, contamination=0.2, random_state=50)
new_df = remove_outliers_isolation_forest(new_df, columns=columns_to_check, contamination=0.1, random_state=50)
```
